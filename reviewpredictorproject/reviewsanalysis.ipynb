```python
def process_size(size_str):
    if size_str.endswith('M'):
        value_str = size_str.replace('M', '')
        return float(value_str) * 1024 if '.' in value_str else int(value_str) * 1024
    elif size_str.endswith('k'):
        return size_str.replace('k', '')
    elif size_str.endswith('+'):
        return size_str.replace('+', '')
    else:
        return int(size_str.replace('Varies with device', '0'))

new_df['Size'] = new_df['Size'].apply(process_size)

        
        

         
    
```
